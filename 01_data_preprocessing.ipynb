```python
# Outliers are below Q1-1.5*IQR or above Q3+1.5*IQR

def find_outliers_IQR(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    IQR = q3 - q1
    outliers = df[((df < (q1 - 1.5*IQR)) | (df > (q3 + 1.5*IQR)))]
    return outliers

# Check outliers for age column
outliers_age = find_outliers_IQR(df['age'])
print("Number of age outliers:", str(len(outliers_age)))
print("Age outliers - min:", outliers_age.min(), " max:", outliers_age.max())
print()

# Check outliers for emplyment_length column
outliers_emp = find_outliers_IQR(df['emplyment_length'])
print("Number of emplyment_length outliers:", str(len(outliers_emp)))
print("emplyment_length outliers - max:", outliers_emp.max())
print()

# Check outliers for loan_interest_rate column
outliers_ir = find_outliers_IQR(df['loan_interest_rate'])
print("Number of loan_interest_rate outliers:", str(len(outliers_ir)))
print("loan_interest_rate outliers - min:", outliers_ir.min())
```

```python
# Find the number of missing values in the dataset
print("Missing values per variable:")
df.isnull().sum()
```

```python
# To find the percentage of missing data per variable
print("Percentage of missing data per variable:")
df.isna().sum()/len(df)*100
```

```python
print("BEFORE Fix 1 — age > 100 erroneous values:")
print(df[df['age'] > 100][['age']])

# Calculate median for imputation
Median_age = df['age'].median()
# Replace impossible values then impute missing
df.loc[df['age'] > 100, 'age'] = Median_age
df['age'] = df['age'].fillna(Median_age)

print("AFTER Fix 1 — age > 100 count:", (df['age'] > 100).sum())
print("Missing age count:", df['age'].isnull().sum())
```

```python
# Replace impossible emplyment_length values (> 60 years = implausible) with median
print("BEFORE Fix 2 — emplyment_length > 60 implausible values:")
print(df[df['emplyment_length'] > 60][['emplyment_length']])

Mean_emplyment_length = df['emplyment_length'].median()
df.loc[df['emplyment_length'] > 60, 'emplyment_length'] = Mean_emplyment_length

print("AFTER Fix 2 — emplyment_length > 60 count:", (df['emplyment_length'] > 60).sum())
```

```python
print("BEFORE Fix 3 — Negative loan_interest_rate:")
print(df[df['loan_interest_rate'] < 0][['loan_interest_rate']])

# Mean of valid (positive) values only
Mean_loan_interest_rate = df.loc[df['loan_interest_rate'] > 0, 'loan_interest_rate'].mean()
df.loc[df['loan_interest_rate'] < 0, 'loan_interest_rate'] = Mean_loan_interest_rate

# Also impute missing loan_interest_rate values
Mean_loan_interest_rate = df['loan_interest_rate'].mean()
df['loan_interest_rate'] = df['loan_interest_rate'].fillna(Mean_loan_interest_rate)

print("AFTER Fix 3 — Negative count:", (df['loan_interest_rate'] < 0).sum())
print("Missing loan_interest_rate count:", df['loan_interest_rate'].isnull().sum())
```

```python
print("BEFORE Fix 4 — Missing payment_default_on_file:", df['payment_default_on_file'].isnull().sum())

# For categorical variable, use mode (most frequent value)
mode_value = df['payment_default_on_file'].mode()[0]
df['payment_default_on_file'] = df['payment_default_on_file'].fillna(mode_value)

print("AFTER Fix 4 — Missing payment_default_on_file:", df['payment_default_on_file'].isnull().sum())
```

```python
# Verify all missing values have been resolved — all should be 0
print("Final verification - :")
df.isna().sum()/len(df)*100
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

# Encode home_ownership
label_encoder = preprocessing.LabelEncoder()
df['home_ownership'] = label_encoder.fit_transform(df['home_ownership'])
print("home_ownership encoding:", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))

# Encode loan_intent
label_encoder = preprocessing.LabelEncoder()
df['loan_intent'] = label_encoder.fit_transform(df['loan_intent'])
print("loan_intent encoding:", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))

# Encode payment_default_on_file
label_encoder = preprocessing.LabelEncoder()
df['payment_default_on_file'] = label_encoder.fit_transform(df['payment_default_on_file'])
print("payment_default_on_file encoding:", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))

df.head()
```
